# Libraries

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, shapiro

# For consistent results
np.random.seed(42)
```

# Non-normal Population

```python
# Generate a non-normal population (Exponential distribution)
population_size = 100000
population = np.random.exponential(scale=2, size=population_size)

# Display basic statistics
print(f"Population Mean: {np.mean(population):.2f}")
print(f"Population Std Dev: {np.std(population):.2f}")
```

# Visualization

```python
plt.figure(figsize=(8, 4))
sns.histplot(population, bins=50, kde=True, color='skyblue')
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Sampling and Sample means

```python
# Sampling parameters
sample_size = 50
num_samples = 1000

# Collect sample means
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

print(f"Sample Means - Mean: {np.mean(sample_means):.2f}, Std Dev: {np.std(sample_means):.2f}")
```

# Visualize Sample Distribution

```python
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=30, kde=True, color='salmon')
plt.title('Sampling Distribution of Sample Mean')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```

# Compare

```python
# Plot sample means with normal curve overlay
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=30, kde=False, stat='density', color='lightgreen')

# Normal curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(sample_means), np.std(sample_means))
plt.plot(x, p, 'k', linewidth=2)

plt.title('Sample Means with Normal Distribution Overlay')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for better visuals
sns.set(style="whitegrid")

# Step 1: Generate a Normal Population
np.random.seed(42)  # for reproducibility
population = np.random.normal(loc=50, scale=10, size=10000)  # mean=50, std=10

# Plot the population distribution
plt.figure(figsize=(8, 4))
sns.histplot(population, bins=30, kde=True, color='skyblue')
plt.title("Population Distribution (Normal)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 2: Draw multiple samples and compute sample means
sample_size = 50
num_samples = 1000
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Step 3: Plot the sampling distribution of the sample mean
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=30, kde=True, color='salmon')
plt.title(f"Sampling Distribution of Sample Mean (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

# Step 4: Print mean and standard deviation
print(f"Population Mean: {np.mean(population):.2f}")
print(f"Population Standard Deviation: {np.std(population):.2f}")
print(f"Sampling Distribution Mean: {np.mean(sample_means):.2f}")
print(f"Sampling Distribution Standard Deviation: {np.std(sample_means):.2f}")
print(f"Theoretical Std Dev of Sample Means: {np.std(population) / np.sqrt(sample_size):.2f}")
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Step 1: Simulate population data
np.random.seed(42)  # reproducibility
population_size = 1000
population_proportion = 0.395

# Simulate population: 1 for smoker, 0 for non-smoker
population = np.random.choice([1, 0], size=population_size, p=[population_proportion, 1 - population_proportion])

# Plot the population distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=population, palette="pastel")
plt.title("Population Smoking Distribution")
plt.xlabel("Smoker (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.xticks([0, 1], ["Non-Smoker", "Smoker"])
plt.show()

# Step 2: Sampling and calculating sample proportions
sample_size = 50
num_samples = 1000
sample_proportions = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    proportion = np.mean(sample)
    sample_proportions.append(proportion)

sample_proportions = np.array(sample_proportions)

# Step 3: Plot sampling distribution of sample proportions
plt.figure(figsize=(8, 4))
sns.histplot(sample_proportions, bins=30, kde=True, color='coral')
plt.title(f"Sampling Distribution of Sample Proportions (n={sample_size})")
plt.xlabel("Sample Proportion of Smokers")
plt.ylabel("Frequency")
plt.show()

# Step 4: Print results
population_prop = np.mean(population)
sample_mean = np.mean(sample_proportions)
sample_std = np.std(sample_proportions)

# Theoretical standard deviation for sample proportions
theoretical_std = np.sqrt(population_prop * (1 - population_prop) / sample_size)

print(f"Population Proportion of Smokers: {population_prop:.3f}")
print(f"Mean of Sample Proportions: {sample_mean:.3f}")
print(f"Standard Deviation of Sample Proportions: {sample_std:.3f}")
print(f"Theoretical Standard Deviation: {theoretical_std:.3f}")
```
