```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([140, 155, 159, 179, 192, 200, 212])
Y = np.array([60, 62, 67, 70, 71, 72, 75])

N = len(X)

sum_x = np.sum(X)
sum_y = np.sum(Y)
sum_xy = np.sum(X * Y)
sum_x2 = np.sum(X ** 2)

m = (N * sum_xy - sum_x * sum_y) / (N * sum_x2 - sum_x ** 2)
c = (sum_y - m * sum_x) / N

Y_pred = m * X + c
ss_total = np.sum((Y - np.mean(Y)) ** 2)  
ss_residual = np.sum((Y - Y_pred) ** 2)  
r_squared = 1 - (ss_residual / ss_total)  


plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', linestyle='-', label='Regression Line')

plt.xlabel("Weight (lbs)")
plt.ylabel("Height (inches)")
plt.title("Simple Linear Regression: Weight vs Height")
plt.legend()
plt.grid()


plt.show()

print(f"Slope (m): {m:.4f}")
print(f"Intercept (c): {c:.4f}")
print(f"Coefficient of Determination (R²): {r_squared:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Given data
X1 = np.array([60, 62, 67, 70, 71, 72, 75, 78]) 
X2 = np.array([22, 25, 24, 20, 15, 14, 14, 11]) 
Y = np.array([140, 155, 159, 179, 192, 200, 212, 215]) 

# Prepare the feature matrix
X = np.column_stack((X1, X2))

# Fit Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Get coefficients
b0 = model.intercept_
b1, b2 = model.coef_

# Print the regression equation
print(f"Regression Equation: Y = {b0:.4f} + {b1:.4f}*X1 + {b2:.4f}*X2")

# Generate points for the regression plane
X1_range = np.linspace(min(X1), max(X1), 10)
X2_range = np.linspace(min(X2), max(X2), 10)
X1_grid, X2_grid = np.meshgrid(X1_range, X2_range)
Y_pred = b0 + b1 * X1_grid + b2 * X2_grid

# Plot the 3D scatter plot and regression plane
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of actual data
ax.scatter(X1, X2, Y, color='red', label='Actual Data')

# Plot regression plane
ax.plot_surface(X1_grid, X2_grid, Y_pred, alpha=0.5, color='cyan')

# Labels and title
ax.set_xlabel("X1 (Feature 1)")
ax.set_ylabel("X2 (Feature 2)")
ax.set_zlabel("Y (Target)")
ax.set_title("Multiple Linear Regression Plane")
ax.legend()
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Given data (Weekly Income X and Weekly Expenditure Y)
X = np.array([80, 100, 120, 140, 160, 180, 200, 220, 240, 260])  # Independent variable
Y = np.array([70, 65, 90, 95, 110, 115, 120, 140, 155, 150])  # Dependent variable

# (i) Estimate the consumption function Y = β0 + β1*X + u
X_const = sm.add_constant(X)  # Add constant for intercept
model = sm.OLS(Y, X_const).fit()  # Fit OLS model
b0, b1 = model.params  # Get coefficients

# (ii) Test the significance of the parameters at 5% level
p_values = model.pvalues  # Extract p-values
summary = model.summary()  # Regression summary

# (iii) Find the coefficient of determination (R²)
r_squared = model.rsquared

# Print results
print(f"Estimated Consumption Function: Y = {b0:.4f} + {b1:.4f} * X")
print(f"Intercept p-value: {p_values[0]:.4f} (Significant if < 0.05)")
print(f"Slope p-value: {p_values[1]:.4f} (Significant if < 0.05)")
print(f"Coefficient of Determination (R²): {r_squared:.4f}")

# Plot regression line
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, b0 + b1 * X, color='red', label='Regression Line')
plt.xlabel("Weekly Income (X)")
plt.ylabel("Weekly Expenditure (Y)")
plt.title("Simple Linear Regression: Income vs Expenditure")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Given data (Total Household Expenditure X and Food Expenditure Y)
X = np.array([100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800])  # Independent variable
Y = np.array([60, 90, 110, 125, 150, 170, 180, 200, 220, 230, 240, 250, 255, 260, 260])  # Dependent variable

# (i) Estimate the expenditure function Y = β0 + β1*X + u
X_const = sm.add_constant(X)  # Add constant for intercept
model = sm.OLS(Y, X_const).fit()  # Fit OLS model
b0, b1 = model.params  # Get coefficients

# (ii) Test the significance of the parameters at 5% level
p_values = model.pvalues  # Extract p-values

# (iii) Find the coefficient of determination (R²)
r_squared = model.rsquared

# Print results
print(f"Estimated Expenditure Function: Y = {b0:.4f} + {b1:.4f} * X")
print(f"Intercept p-value: {p_values[0]:.4f} (Significant if < 0.05)")
print(f"Slope p-value: {p_values[1]:.4f} (Significant if < 0.05)")
print(f"Coefficient of Determination (R²): {r_squared:.4f}")

# Plot regression line
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, b0 + b1 * X, color='red', label='Regression Line')
plt.xlabel("Total Household Expenditure (X)")
plt.ylabel("Food Expenditure (Y)")
plt.title("Simple Linear Regression: Household vs Food Expenditure")
plt.legend()
plt.grid()
plt.show()
```
